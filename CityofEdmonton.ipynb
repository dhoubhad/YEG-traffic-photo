```python
map_yeg = folium.Map(location=[53.55, -113.55], 
                     zoom_start=10, 
                     tiles = 'Stamen Toner')
#folium.Marker([53.55, -113.55]).add_to(map_yeg)
```

```python
def getdata():
    api = 'https://data.edmonton.ca/resource/rgpv-ugyr.json'
    req = requests.get(api)
    df = pd.DataFrame(req.json())
    return df
```

```python
df = getdata()
```

```python
def extract_coordinates():
    df2 = df['the_geom'].apply(pd.Series)
    df.drop(columns = ['the_geom'])
    df2 = df2['coordinates'].apply(pd.Series)
    df2.columns = ['GPS']
    return df.join(df2)
```

```python
df = extract_coordinates()
df.head()
```

```python
for index, row in df.iterrows():
    coord = pd.DataFrame(row['GPS'])
    coord.columns = ['Longitude', 'Latitude'] 
    for index, row in coord.iterrows():
        folium.Marker([row['Latitude'], row['Longitude']]).add_to(map_yeg)  
```
